Notebook to demonstrate simple low pass filtering of 1D signal

```python
import numpy as np
import math as m
import matplotlib.pyplot as plt
import scipy.fftpack as fft
import cv2
import util
import scipy.fftpack as fftp
import matchedFilter as mF
import scipy as sp
```

```python
# for Principle Component Analysis
from sklearn.decomposition import PCA
```

```python
# number of sample points
N = 300
# spacing
space = 1.0 / 100.0
x = np.linspace(0, N*space, N)
print len(x)
y = np.sin(2.0 * np.pi * x) + 0.25 * np.sin(15 * 2.0 * np.pi * x)
```

```python
plt.figure()
plt.plot(x,y)
plt.title('Raw Signal')
```

```python
# trying shift now
#y = fft.fftshift(y)
#plt.plot(x,y)
```

```python
trans = fft.fft(y)
#print len(trans)
xF = np.linspace(-1. / (2 * space), 1. / (2*space), N)
```

```python
# trying shift on trans
trans = fft.fftshift(trans)
```

```python
plt.figure()
plt.plot(xF, 2. / N * np.abs(trans))
plt.title('Transformed Signal')
plt.show()
```

### Applying low-pass filter

```python
# anything higher than number of oscillations, we get rid of
numOsc = 10

lp = trans.copy()
#print np.logical_and(xF > 10, xF < 90) 
lp[np.logical_or(xF > 10, xF < -10) ] = 0
```

```python
plt.figure()
plt.plot(xF, 2. / N * np.abs(lp))
plt.title('Filtered Signal')
plt.show()
```

### Take Inverse FT

```python
shift = fft.ifftshift(lp)
inv = fft.ifft(shift)
```

```python
plt.figure()
plt.plot(x,inv)
plt.title("Filtered Signal")
```

# Trying High Pass Filter Now

```python
hp = trans.copy()
#print np.concatenate((np.where(xF<10)[0],np.where(xF>90)[0]))
#bandPass = np.multiply(np.where(xF < 10),np.where(xF> 90))
#hp[np.concatenate((np.where(xF<10)[0],np.where(xF>90)[0]))] = 0
hp[np.logical_and(xF<10,xF>-10)] = 0
```

```python
plt.figure()
plt.plot(xF, 2. / N * np.abs(hp))
plt.title('Filtered Signal')
plt.show()
```

```python
shift = fft.ifftshift(hp)
inv = fft.ifft(shift)
plt.plot(x,inv)
```


### Example 1D Matched Filtering

```python
plt.figure()
plt.plot(trans)
plt.show()

filtFFT = fft.fft(inv)
filtFFT = fft.fftshift(filtFFT)
matchedFilt = trans * filtFFT
matchedFilt = fft.ifftshift(matchedFilt)
matchedFilt = fft.ifft(matchedFilt)
plt.figure()
plt.plot(x,matchedFilt)
plt.show()
```

### Trying 2D Filtering

```python
def givePS(array):
    '''
    Function to give the power spectrum of an array
    and plot the log transform of it
    '''
    dims = np.shape(array)
    psd = np.absolute(array)
    
    ### cut off all ridiculously low signal that is likely artifacts
    psd[psd<0.0002] = 0
    
    plt.figure()
    plt.imshow(np.log(np.real(psd.copy())))
    plt.title("Log Transform of the Real Component of the Power Spectrum")
    plt.colorbar()
    plt.show()
    return psd
```

#### 2D Example Using Simple WT representation

Construct WT image

```python
wtImg = np.zeros((50,100),float)

thickness = 3
space = 12
position = 0

while position < 99:
    position += space
    wtImg[:,position-thickness:position+thickness] = 1.


### apply smoothing to make the WT represention more realistic
kernel = np.ones((3,3),float) / 9.
wtImg = mF.matchedFilter(wtImg,kernel,demean=False)


### now lets add in some background noise 
wtImg -= 0.25
wtImg = wtImg + 0.25 * np.random.rand(50,100)

    
plt.figure()
plt.imshow(wtImg,cmap='gray')
plt.title("WT Representation")
plt.colorbar()
plt.show()
```
