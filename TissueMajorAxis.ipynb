```python
import numpy as np
import math as m
import matplotlib.pyplot as plt
import scipy.fftpack as fftp
import cv2
import util
import myocyteFigs as mF
import copy
import imutils
import tissue

plt.rcParams['figure.figsize'] = [16,9]
```

```python
class empty:pass
cases = dict()
```

```python
def highpass(transform,threshold):
    filtered = transform.copy()
    filtered[filtered < threshold] = 0
    return filtered
```

```python
def givePS(array):
    dims = np.shape(array)
    for row in range(dims[0]):
        for col in range(dims[1]):
            array[row,col] = np.linalg.norm(array[row,col])
    return array
```

```python
case=empty()
# want 200x200 uM subsection in WT region
case.loc_um = [3923,0] # need to define from upper-left corner of ROI in fiji
case.extent_um = [200,200]

cases['wtlike'] = case
```

```python
tissue.SetupCase(cases['wtlike'])
```

```python
tissue.SetupFilters()
paramDict = tissue.SetupParams()
print paramDict['snrThresh']
results = tissue.Run(case,paramDict=paramDict)
```

```python
img = case.subregion.copy().astype(float)
img /= np.max(img)

img = util.PadWithZeros(img, 200)
img = imutils.rotate(img, -30)

dims = np.shape(img)

imshow(img,cmap='gray')
colorbar()
```

### Create 2D Hamming Window

```python
alpha = .53836
beta = .46164
Wrow = np.asarray([alpha - beta * np.cos(2*np.pi*n/(dims[0]-1)) for n in range(dims[0])])
Wcol = np.asarray([alpha - beta * np.cos(2*np.pi*n/(dims[1]-1)) for n in range(dims[1])])

window = np.outer(Wrow,Wcol)

figure()
imshow(window)
colorbar()
show()
```

```python
windowedImg = img * window
imshow(windowedImg)
colorbar()
```

```python
#sdimg = fftp.fftshift(img)
sdimg = img.copy()
#winsdimg = fftp.fftshift(windowedImg)
winsdimg = windowedImg.copy()


transform1 = fftp.fft2(sdimg)
transform2 = fftp.fft2(winsdimg)
```

```python
#new1 = fftp.ifftshift(transform1.copy())
new1 = fftp.fftshift(transform1.copy())
PS1 = np.real(givePS(new1.copy()))

new2 = fftp.fftshift(transform2.copy())
PS2 = np.real(givePS(new2.copy()))
```

```python
figure()
f,axarr = plt.subplots(2,2)
axarr[0,0].imshow(img,cmap='gray')
axarr[0,0].set_title("Raw Image")
axarr[0,1].imshow(windowedImg)
axarr[0,1].set_title("Image with Window Applied")
axarr[1,0].imshow(np.log(PS1))
axarr[1,0].set_title("Log Transform of Power Spectrum w/o Window")
axarr[1,1].imshow(np.log(PS2))
axarr[1,1].set_title("Log Transform of Power Spectrum w/ Window")
show()
```

### Create Gaussian Fourier Domain Mask

```python
sigma = 25
import scipy as sp
gaussMask = sp.signal.gaussian(dims[1],sigma)
gaussMask /= np.max(gaussMask)


plt.plot(gaussMask)
plt.ylim([0,1])
```

```python
mask = np.zeros_like(img)
for row in range(np.shape(img)[0]):
    mask[row,:] = gaussMask
imshow(mask)
colorbar()
```

### Multiply Fourier Transform by Mask

```python
#masked = np.multiply(new1, mask)
masked = np.multiply(new1, mask)
real = np.real(givePS(masked.copy()))

imshow(np.log(real))
colorbar()
```

```python
ishift = fftp.ifftshift(masked)
h = np.real(fftp.ifft2(ishift))
#h = fftp.ifftshift(h)
```

```python
f, (ax1,ax2) = plt.subplots(1,2)
ax1.imshow(img,cmap='gray')
ax1.set_title("Rotated Raw Image")
ax2.imshow(h)
ax2.set_title("Inverse FT Image")
```

```python
offset = 80
sigma = 35
print sigma

domain = range(0, dims[1], 1)
#print domain
gauss = lambda x: (1./(sigma*sqrt(2*np.pi))) * \
        ( exp(-0.5 * (float((x-(int(float(len(domain))/2.)+offset)))/sigma)**2) + \
         exp(-0.5 * (float((x-(int(float(len(domain))/2.)-offset)))/sigma)**2) \
         )
         #+ exp(-0.5 * (float((x-(int(float(len(domain))/2.)+2*offset)))/sigma)**2) + \
         #exp(-0.5 * (float((x-(int(float(len(domain))/2.)-2*offset)))/sigma)**2) )
func = np.asarray([gauss(x) for x in domain])
func /= np.max(func)
    
plt.plot(domain, func)
```

```python
bimodal = np.zeros_like(img)
for row in range(np.shape(img)[0]):
    bimodal[row,:] = func
imshow(bimodal)
colorbar()
```

```python
masked = np.multiply(new1, bimodal)
ishift = fftp.ifftshift(masked)
h = np.real(fftp.ifft2(ishift))
#h = fftp.ifftshift(h)
```
