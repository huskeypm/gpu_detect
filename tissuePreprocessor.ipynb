### Endgoal: Function to preprocess the entire tissue level image so no further preprocessing is necessary for any of the figure generation routines (aside from the reorientation)



```python
def showSubsection(img):
    '''
    Function to show a small subsection of the entire tissue image for debugging purposes
    '''
    plt.figure()
    plt.imshow(img[10000:10500,10000:10500],cmap='gray')
    plt.colorbar()
    plt.title("Subsection of tissue image")
    plt.show()
    
    print "Shape of tissue image:", np.shape(img)
```

```python
def saveNewTissueImg(img):
    '''
    Function to save the new tissue level image for DISPLAY PURPOSES.
    
    This means the original dynamic range is not preserved as it will have to be widened to display on a normal screen
    '''
    
    #newImg = img.copy().astype(np.float32)
    #newImg *= 255. / np.max(newImg)
    #newImg = np.round(newImg).astype(np.uint8)
    
    newImg = img
    
    cv2.imwrite('displayTissue.png',newImg)
    
    print "Wrote tissue image for display purposes as 'displayTissue.png'"
```

#### Load in tissue level image

```python
root = "/net/share/pmke226/DataLocker/cardiac/Sachse/171127_tissue/"

fileName = "tissue.tif"

tissueImg = cv2.imread(root+fileName)
tissueImg = cv2.cvtColor(tissueImg, cv2.COLOR_BGR2GRAY)

showSubsection(tissueImg)
```

#### Rescale Tissue Level Image

```python
## measured previously
imgTwoSarcSize = 22
## specified with filter generation
filterTwoSarcSize = 25
scale = float(filterTwoSarcSize) / float(imgTwoSarcSize)

resizedTissueImg = cv2.resize(tissueImg,None,fx=scale,fy=scale,interpolation=cv2.INTER_CUBIC)

showSubsection(resizedTissueImg)
```

#### Experimenting with Clipping the membrane off at a specified value before CLAHE

```python
## TTs seem to have fluorescent value ~7, so can we tease this out by clipping the membrane at 7?
clipImg = False
if clipImg:
    ceilingValue = 7
    clippedTissueImg = resizedTissueImg.copy()
    clippedTissueImg[clippedTissueImg > ceilingValue] = ceilingValue
else:
    clippedTissueImg = resizedTissueImg.copy()

showSubsection(clippedTissueImg)
```

#### Apply Different, more global CLAHE routine

This time we're opting to perform a CLAHE across 1/8th of the image (empirical) instead of doing CLAHE at the level of the sarcomere. This should reduce the amount of noise that is getting enhanced

```python
tissueDims = np.shape(clippedTissueImg)

claheTileSize = int(1./8. * tissueDims[0])

CLAHEDtissueImg = pp.applyCLAHE(clippedTissueImg, claheTileSize)

showSubsection(CLAHEDtissueImg)
```

#### Taking a look at the new tissue level image to see how the clipping changed the CLAHE

```python
saveNewTissueImg(CLAHEDtissueImg)
```

Looks like the CLAHE did well. Now we need to clip the bottom of the image out since the CLAHE increased the background intensity

```python
floorValue = 6
ceilingValue = 10
clippedTissueImg = CLAHEDtissueImg.copy()
clippedTissueImg[clippedTissueImg < floorValue] = floorValue
clippedTissueImg[clippedTissueImg > ceilingValue] = ceilingValue
clippedTissueImg -= floorValue
clippedTissueImg = clippedTissueImg.astype(np.float32)
clippedTissueImg *= 255. / np.max(clippedTissueImg)
clippedTissueImg = clippedTissueImg.astype(np.uint8)

#saveNewTissueImg(clippedTissueImg)
cv2.imwrite("./myoimages/preprocessedTissue.png",clippedTissueImg)
```
