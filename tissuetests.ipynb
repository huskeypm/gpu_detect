#### Functions for rescaling 'fiji' coordinates for image to those used by cv2

```python

#print dim
#print px_per_um

def conv_fiji(x_um,y_um): # in um
    y_px = int(x_um * params.px_per_um[1])
    x_px = int(y_um * params.px_per_um[0])
    return x_px,y_px

def get_fiji(gray, loc_um,d_um):
    loc= conv_fiji(loc_um[0],loc_um[1])
    d = conv_fiji(d_um[0],d_um[1])
    subregion = gray[loc[0]:(loc[0]+d[0]),loc[1]:(loc[1]+d[1])]
    return subregion

```

### Create lobe filter
- The point of this filter is to penalize 'side lobe' responses from the MF, which would typically arise from signals not resembling the MF. 
- This is an ad hoc approach that is best informed from running the MF first, assessing the correlation plane, and discounting the corr accordingly

Further down in the notebook I determined that the 'correct' signal generates corr. responses that are about 4px wide, while 4px to either side of that response should be 'side lobes'


```python
# test with embedded signal
vert = int(3.5 * params.px_per_um[0])
sigW = 4  # px
lobeW = 4 # px
lobemf = np.ones([vert,lobeW + sigW + lobeW])
lobemf[:,lobeW:(lobeW+sigW)]=0.

#imshow(mf,cmap='gray')

import imutils
lobemfr = imutils.rotate_bound(lobemf,-rot)
imshow(lobemfr,cmap='gray')

lobemfr = makeSquare(lobemfr)
imshow(lobemfr,cmap='gray')
```

```python
import matchedFilter as mF
```

```python

def docalc(img,mf,lobemf=None,corrThresh=0.,name="corr.png"):
    results = empty()
    
    # get correlation plane 
    corr = mF.matchedFilter(img,mf,parsevals=False,demean=False)
    subplot(2,2,1)
    imshow(img,cmap='gray')
    #plt.colorbar()

    subplot(2,2,3)
    imshow(corr)
    #plt.colorbar()
    plt.gcf().savefig("x.png",dpi=300)

    corrThreshed = np.zeros_like(corr)
    corrThreshed[where(corr>corrThresh)] = 1.
    
    # get side lobe penalty
    if isinstance(lobemf,np.ndarray):
        corrlobe = mF.matchedFilter(corr,lobemf,parsevals=True,demean=False)
        
    else:    
        corrlobe = np.ones_like(corr)
        
    # zero out subthresh components
    subplot(2,2,4)
    imshow(corrlobe)
    
    snr = corr/corrlobe * corrThreshed
    #snr = corrThreshed

    subplot(2,2,2)
    imshow(snr)
    #plt.colorbar()
    plt.gcf().savefig(name,dpi=300)

    
    results.img = img
    results.corr = corr
    results.corrlobe = corrlobe
    results.snr = snr
    
    return results
```
